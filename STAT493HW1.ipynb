```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

#Formula 1 2022 Season

```python
url = "https://www.skysports.com/f1/standings"
R = requests.get(url)
```

```python
soup = BeautifulSoup(R.text, "html.parser")
```

```python
l_table = soup.find("table", class_ = "standing-table__table")
```

```python
for driver in l_table.find_all("tbody"):
  rows = driver.find_all("tr")
  f1_drivers = []
  drivers_nat = []
  drivers_team = []
  drivers_points = []
  for row in rows:
    f1_drivers.append(row.find("td", class_ = "standing-table__cell standing-table__cell--name").text.strip())
    drivers_nat.append(row.find_all("td", class_ = "standing-table__cell")[2].text.strip())
    drivers_team.append(row.find_all("td", class_ = "standing-table__cell")[3].text.strip())
    drivers_points.append(row.find_all("td", class_ = "standing-table__cell")[4].text.strip())

data = pd.DataFrame({"Drivers Name":f1_drivers,
                    "Nationality":drivers_nat,
                    "Team":drivers_team,
                    "Points":drivers_points})
data.style.set_caption("2022 Season")
```

# Formula 1 2021 Season

```python
url = "https://www.skysports.com/f1/stats/2021/races"
R = requests.get(url)
```

```python
soup = BeautifulSoup(R.text, "html.parser")
```

```python
l_table = soup.find("table", class_ = "table -small -nobordertop ")
```

```python
for driver in l_table.find_all("tbody"):
  rows = driver.find_all("tr")
  circuit = []
  gp_date = []
  gp_winner = []
  drivers_team = []
  best_time = []
  pole = []
  for row in rows:
    circuit.append(row.find("td", class_ = "strap3 f1-stats-table__participant").text.strip())
    gp_date.append(row.find_all("td")[1].text.strip())
    gp_winner.append(row.find_all("td")[2].text.strip())
    drivers_team.append(row.find_all("td")[3].text.strip())
    best_time.append(row.find_all("td")[4].text.strip())
    pole.append(row.find_all("td")[5].text.strip())


data = pd.DataFrame({"Grand Prix":circuit,
                    "Date":gp_date,
                    "Winner":gp_winner,
                    "Team":drivers_team,
                    "Time":best_time,
                    "Pole":pole})
data.style.set_caption("2021 Season")
```
